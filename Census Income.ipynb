```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in df.columns:
    if df[col].dtypes == 'object':
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])
```

```python
X = df.drop('Income', axis=1)
Y = df['Income']
```

```python
from sklearn.ensemble import ExtraTreesClassifier
selector = ExtraTreesClassifier(random_state=45)
```

```python
selector.fit(X, Y)
```

```python
feature_imp = selector.feature_importances_
for index, val in enumerate(feature_imp):
    print(index, round((val * 100), 2))
```

```python
X.info()
```

```python
X = X.drop(['Workclass', 'Education', 'Race', 'Sex','Capital_loss', 'Native_country'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
for col in X.columns:
    scaler = StandardScaler()
    X[col] = scaler.fit_transform(X[col].values.reshape(-1, 1))
```

```python
X[col]
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=45)
```

```python
ros.fit(X, Y)
```

```python
X_resampled, Y_resampled = ros.fit_resample(X, Y)
```

```python
round(Y_resampled.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
#Logistic Regression
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_resampled, Y_resampled, test_size=0.2, random_state=45)
```

```python
"X_train shape:", X_train.shape
```

```python
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=45)
```

```python
log_reg.fit(X_train, Y_train)
Y_pred_log_reg = log_reg.predict(X_test)
```

```python
Y_pred_log_reg
```

```python
#Support Vector Classifier
```

```python
from sklearn.svm import SVC
svc = SVC(random_state=45)
```

```python
svc.fit(X_train, Y_train)
Y_pred_svc = svc.predict(X_test)
```

```python
Y_pred_svc
```

```python
#Decision Tree Classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree = DecisionTreeClassifier(random_state=45)
```

```python
dec_tree.fit(X_train, Y_train)
```

```python
Y_pred_dec_tree = dec_tree.predict(X_test)
```

```python
Y_pred_dec_tree
```

```python
#Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
ran_for = RandomForestClassifier(random_state=45)
```

```python
ran_for.fit(X_train, Y_train)
```

```python
Y_pred_ran_for = ran_for.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
print('Logistic Regression:')
print('Accuracy score:', round(accuracy_score(Y_test, Y_pred_log_reg), 2))
print('F1 score:', round(f1_score(Y_test, Y_pred_log_reg), 2))
```

```python
print('Support Vector Classifier:')
print('Accuracy score:', round(accuracy_score(Y_test, Y_pred_svc), 2))
print('F1 score:', round(f1_score(Y_test, Y_pred_svc), 2))
```

```python
print('Decision Tree Classifier:')
print('Accuracy score:', round(accuracy_score(Y_test, Y_pred_dec_tree), 2))
print('F1 score:', round(f1_score(Y_test, Y_pred_dec_tree), 2))
```

```python
print('Random Forest Classifier:')
print('Accuracy score:', round(accuracy_score(Y_test, Y_pred_ran_for), 2))
print('F1 score:', round(f1_score(Y_test, Y_pred_ran_for), 2))
```
