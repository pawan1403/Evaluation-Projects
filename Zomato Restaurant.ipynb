```python
Acft = df[df['Average_Cost_for_two'] > 200000]
Acft = Acft.sort_values(by='Average_Cost_for_two')
plt.figure(figsize = (8,6))
sns.barplot(data = Acft, x = "Cuisines", y = "Average_Cost_for_two", palette = "Purples")
plt.xticks(rotation = 90)
plt.title("Cuisines_with_Average_Cost > 200000")
plt.xlabel("Cuisines")
plt.ylabel("Average_Cost_for_two")
```
